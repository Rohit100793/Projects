```python
def myplot(data,playerlist = Players):
    Col = {"KobeBryant":'Red',"JoeJohnson":'Blue',"LeBronJames":'Green',"CarmeloAnthony":'Orange',"DwightHoward":'Yellow',"ChrisBosh":'Magenta',"ChrisPaul":'Pink',"KevinDurant":'cyan',"DerrickRose":'Black',"DwayneWade":'Brown'}
    Marker = {"KobeBryant":'o',"JoeJohnson":'8',"LeBronJames":'s',"CarmeloAnthony":'p',"DwightHoward":'*',"ChrisBosh":'^',"ChrisPaul":'X',"KevinDurant":'d',"DerrickRose":'.',"DwayneWade":'H'}
    for name in playerlist:
        plt.plot(data[Pdict[name]], c=Col[name], marker=Marker[name], ls='--', label=name)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons)
    plt.show()
```
