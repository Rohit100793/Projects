Creating a scatter plot to depict the Birth Rate vs Life Expectancy of the years 1960 and 2013. This graph will be categorised by Country regions which will give a clear indication which region has a better Life Expectancy/Higher Biirth Rate etc. Let's have a look below.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
visual_1960 = sns.lmplot(data = merged_data1, x = 'Birthrate', y = 'LifeExp1960', fit_reg = False, hue = 'CountryRegion', size=8,\
                    scatter_kws={"s": 70})
```

In 1960 the Life expectancy was was greatest in Europe region whereas Birth Rate was hovering nearly around 12. In Asia the Life expectancy was nearly between 52 to 75 and Birth rate was mostly around 25.
Similarly in America the life expectancy was around 65 where birth rate was around 15 whereas in Africa it was between 30 to 40 and birth rate was around 35.

```python
visual_2013 = sns.lmplot(data = merged_data1, x = 'Birthrate', y = 'LifeExp2013', fit_reg = False, hue = 'CountryRegion', size=8,
                    scatter_kws={"s": 70})
```

In 2013 life expectancy increase in Europe to around 83 whereas in America it increased to 77. In Asia the life expectancy hover around the region 67-68 and in Africa it increased to around the region 55-65 from 30-40.
